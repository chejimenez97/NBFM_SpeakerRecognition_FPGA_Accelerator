## Generate Vivado Project

At this point, the needed tcl scripts should have been loaded into target folder.

run design.tcl script to generate the vivado project in command line:
`vivado -mode batch -source design.tcl `

add the interconnect pragmas at firmware/myproject.cpp

`#pragma HLS INTERFACE axis port=input_1`

`#pragma HLS INTERFACE axis port=layer23_out`

`#pragma HLS INTERFACE s_axilite port=return bundle=control`


delete /impl directory to force vivado to generate new files.

recompile and re-export the IP using the build_prj.tcl:
`vitis_hls -f build_prj.tcl &> ./ip_repo/build_log.txt `

open the vivado project and add the compiled IP. Typically, it will be exported at:
myproject_prj/solution1/impl/ip

Import IP at vivado.

add myproject IP, axi stream width converter the DMA and the zynq.

Connect the inputs and outputs manually:
myproject_out        -> s_axis converter
M_axis converter    -> s_axis_s2mm
s_axis_mm2s         -> myproject_input

set ps-pl clocks at ip config
Slave axi Hp -> HP0 PFD

set output PL clocks
Select zynq_ultra_ps_e_0 → Run Block Automation → enable FCLK0 (or enable PL fabric clocks) and finish.
In BD canvas select the AXI DMA → Run Connection Automation or manually connect its clock pins (aclk, m_axi_sg_aclk, m_axi_mm2s_aclk) to zynq_ultra_ps_e_0/FCLK_CLK0.

Run connection automation

generate wrapper and define top module

validate and synthesize
